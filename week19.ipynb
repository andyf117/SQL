```python
## ***** ANSWER ******
##### Data Analysis
# 1. List the following details of each employee: employee number, last name, first name, gender, and salary.
#####

pd.read_sql_query(""" SELECT e.emp_no, e.last_name, e.first_name, e.gender, s.salary
 FROM "Employees" as e 
 LEFT JOIN "Salaries" s
 on e.emp_no = s.emp_no  """,conn)
```

```python
## ***** ANSWER ******
##### Data Analysis
# 2. List employees who were hired in 1986.
#####
pd.read_sql_query(""" SELECT e.emp_no, e.last_name, e.first_name, e.gender, s.salary, e.hire_date
 FROM "Employees" as e 
 LEFT JOIN "Salaries" s
 on e.emp_no = s.emp_no  WHERE extract(year from e.hire_date) = 1986 """,conn)
```

```python
## ***** ANSWER ******
##### Data Analysis
# 3.  List the manager of each department with the following information: 
#     department number, department name, the manager's employee number,
#     last name, first name, and start and end employment dates..
#####
pd.read_sql_query(""" SELECT d.dept_no AS "Department Number", d.dept_name AS "Department Name", 
dm.emp_no AS "Employee Number",
e.last_name AS "Lastname", e.first_name AS "Firstname",
dm.from_date AS "Start Employment Date", 
dm.to_date AS "End Employment Date"  
FROM "Department" d LEFT JOIN "Dept_Manager" dm 
ON dm.dept_no = d.dept_no LEFT JOIN "Employees" e 
ON e.emp_no = dm.emp_no order by e.last_name """,conn)
```

```python
## ***** ANSWER ******
##### Data Analysis
# 4. List the department of each employee with the 
#     following information: employee number, last name,
#     first name, and department name.
#########

pd.read_sql_query(""" SELECT d.dept_no AS "Department Number", d.dept_name AS "Department Name", 
e.emp_no AS "Employee Number",
e.last_name AS "Lastname", e.first_name AS "Firstname"
FROM "Department" d LEFT JOIN "Dept_Emp" de 
ON de.dept_no = d.dept_no LEFT JOIN "Employees" e 
ON e.emp_no = de.emp_no order by e.last_name """,conn)

```

```python
## ***** ANSWER ******
##### Data Analysis
# 5. List all employees whose first name is "Hercules" and last names begin with "B."
#####


pd.read_sql_query(""" SELECT e.emp_no, e.last_name, e.first_name, e.gender, s.salary
 FROM "Employees" as e 
 LEFT JOIN "Salaries" s
 on e.emp_no = s.emp_no WHERE e.first_name = 'Hercules' 
 AND e.last_name LIKE 'B%' """,conn)
```

```python
## ***** ANSWER ******
##### Data Analysis
# 6. List all employees in the Sales department, including their employee number, last name, first name, and department name.
#####


pd.read_sql_query(""" SELECT d.dept_no AS "Department Number", d.dept_name AS "Department Name", 
e.emp_no AS "Employee Number",
e.last_name AS "Lastname", e.first_name AS "Firstname"
FROM "Department" d LEFT JOIN "Dept_Emp" de 
ON de.dept_no = d.dept_no LEFT JOIN "Employees" e 
ON e.emp_no = de.emp_no where de.dept_no = 'd007' """,conn)
```

```python
## ***** ANSWER ******
##### Data Analysis
# 7. List all employees in the Sales and Development departments, 
#    including their employee number, last name, first name, and department name.
#####


pd.read_sql_query(""" SELECT d.dept_no AS "Department Number", d.dept_name AS "Department Name", 
e.emp_no AS "Employee Number",
e.last_name AS "Lastname", e.first_name AS "Firstname"
FROM "Department" d LEFT JOIN "Dept_Emp" de 
ON de.dept_no = d.dept_no LEFT JOIN "Employees" e 
ON e.emp_no = de.emp_no where de.dept_no = 'd007' or de.dept_no = 'd005' """,conn)
```

```python
## ***** ANSWER ******
##### Data Analysis
# 8. In descending order, list the frequency count of employee last names, i.e., how many employees share each last name
#####


pd.read_sql_query(""" SELECT COUNT(*) as total_count ,
e.last_name FROM "Employees" e 
GROUP BY e.last_name ORDER BY 
e.last_name DESC """,conn)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Preparing for plotting
titles_with_avg_salary = pd.read_sql_query(""" 
SELECT  MIN(t.title) as "Title",
AVG(s.salary) as "Average Salary",
COUNT(*) as "Total Employees"
FROM "Titles" t
LEFT JOIN "Salaries" s
ON t.emp_no = s.emp_no 
GROUP BY t.title    """,connection)

titles_with_avg_salary_plot = titles_with_avg_salary.plot(kind='bar', figsize=(20,6),x='Title',y='Average Salary')
titles_with_avg_salary_plot.set_xlabel("Position Titles")

plt.title("Average Salary for Various Titles")
plt.show()
plt.tight_layout()
```
